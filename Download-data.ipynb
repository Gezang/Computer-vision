# To download the dataset for the project, you can run the following


```python
import shutil
from pathlib import Path
import zipfile

import kagglehub

dataset_ref = "masoudnickparvar/brain-tumor-mri-dataset"

download_path = Path(kagglehub.dataset_download(dataset_ref))
project_root = Path.cwd()
target_dir = project_root / "archive"

if target_dir.exists():
    shutil.rmtree(target_dir)
target_dir.mkdir(parents=True, exist_ok=True)

zip_files = [p for p in download_path.iterdir() if p.suffix == ".zip"]

if zip_files:
    for zip_path in zip_files:
        with zipfile.ZipFile(zip_path) as zip_file:
            zip_file.extractall(target_dir)
else:
    for item in download_path.iterdir():
        destination = target_dir / item.name
        if item.is_dir():
            shutil.copytree(item, destination)
        else:
            shutil.copy2(item, destination)

training_dir = target_dir / "Training"
if not training_dir.exists():
    child_dirs = [p for p in target_dir.iterdir() if p.is_dir()]
    if len(child_dirs) == 1 and (child_dirs[0] / "Training").exists():
        nested_root = child_dirs[0]
        for child in nested_root.iterdir():
            shutil.move(str(child), str(target_dir / child.name))
        nested_root.rmdir()

print(f"Dataset cached at: {download_path}")
print(f"Dataset ready under: {target_dir}")
```
